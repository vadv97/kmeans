```python
import numpy as np
import pandas as pd
import datetime
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt, numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import AgglomerativeClustering
from matplotlib.colors import ListedColormap
from sklearn import metrics
import warnings
import sys
if not sys.warnoptions:
    warnings.simplefilter("ignore")
np.random.seed(42)
```

```python
#Loading dataset 

data = pd.read_csv("marketing_campaign.csv", sep="\t")
print("Number of datapoints:", len(data))
data.head()
```

```python
data.info()
```

```python
#Dropping null values from the dataset

data = data.dropna()
print("The total number of data-points after removing the rows with missing values are:", len(data))
```

```python
data["Dt_Customer"] = pd.to_datetime(data["Dt_Customer"])
dates = []
for i in data["Dt_Customer"]:
    i = i.date()
    dates.append(i)  
#Dates of the newest and oldest recorded customer
#print("newest customer's enrolment date:",max(dates))
#print("oldest customer's enrolment date:",min(dates))
```

```python
##Customer_For = the number of days the customers started to shop relative to last recorded date
```

```python
days = []
d1 = max(dates) #taking it to be the newest customer
for i in dates:
    delta = d1 - i
    days.append(delta)
data["CustomerFor"] = days
data["CustomerFor"] = pd.to_numeric(data["CustomerFor"], errors="coerce")
```

```python
print("Total categories in Marital_Status:\n", data["Marital_Status"].value_counts(), "\n")
print("Total categories in Education:\n", data["Education"].value_counts())
```

```python
#Feature Engineering

#Age of customer today 
data["Age"] = 2022-data["Year_Birth"]

#Total spendings on various items
data["Spent"] = data["MntWines"]+ data["MntFruits"]+ data["MntMeatProducts"]+ data["MntFishProducts"]+ data["MntSweetProducts"]+ data["MntGoldProds"]

#Deriving living situation by marital status "Alone"
data["Living_With"]=data["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})

#Feature indicating total children living in the household
data["Children"]=data["Kidhome"]+data["Teenhome"]

#Feature for total members in the householde
data["Family_Size"] = data["Living_With"].replace({"Alone": 1, "Partner":2})+ data["Children"]

#Feature pertaining parenthood
data["Is_Parent"] = np.where(data.Children> 0, 1, 0)

#Segmenting education levels in three groups
data["Education"]=data["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})

#For clarity
data=data.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})

#Dropping some of the redundant features
to_drop = ["Marital_Status", "Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID"]
data = data.drop(to_drop, axis=1)
```

```python
#Plotting the data 

sns.set(rc={"axes.facecolor":"#FEF9EF","figure.facecolor":"#FFF9ED"})
pallet = ["#000D6B", "#9C19E0", "#FF5DA2", "#99DDCC", "#9F8A78", "#F3AB60"]
cmap = colors.ListedColormap(["#682F2F", "#9E726F", "#D6B2B1", "#B9C0C9", "#9F8A78", "#F3AB60"])

#Plotting some ket features
To_Plot = [ "Income", "Recency", "CustomerFor", "Age", "Spent", "Is_Parent"]
print("Plot Of the Features: Subset")
plt.figure()
sns.pairplot(data[To_Plot], hue= "Is_Parent",palette= (["#8946A6","#CCD1E4"]))


#Taking hue 
plt.show()
```

```python
#Dropping the outliers by setting a cap on Age and income. 

data = data[(data["Age"]<90)] # data is a little old hence we wanted to cap the cases of probably unusable datapoints
data = data[(data["Income"]<600000)]
print("The total number of data-points after removing the outliers are:", len(data))
```

```python
#correlation matrix

corrmat= data.corr()
plt.figure(figsize=(20,20))  
sns.heatmap(corrmat,annot=True, cmap=cmap, center=0)
```

```python
#Data Preprocessing

#Listing categorical variables

s = (data.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:", object_cols)
```

```python
#Label encoding the object dtypes

LE=LabelEncoder()
for i in object_cols:
    data[i]=data[[i]].apply(LE.fit_transform)
    
print("All features are now numerical")
```

```python
#Creating a copy

ds = data.copy()


# creating a subset of dataframe by dropping the features on deals accepted and promotions

cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
ds = ds.drop(cols_del, axis=1)

#Scaling using StandardScaler

scaler = StandardScaler()
scaler.fit(ds)
scaled_ds = pd.DataFrame(scaler.transform(ds),columns= ds.columns )
print("All features are now scaled")
```

```python
#Reducing the dimensionality using the scaled data

print("New Dataframe for further modelling:")
scaled_ds.head()
```

```python
##Dimensionality reduction with PCA

#PCA to reduce dimensions to 3

pca = PCA(n_components=3)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
#3D Projection Of Data with reduced dimensions

x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]

#Plotting data

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="violet", marker="o" )
ax.set_title("3D Plot of Data ")
plt.show()
```

```python
#Clustering

#Elbow method to find ideal numbers of clusters

print('Elbow Method to determine the number of clusters:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_ds)
Elbow_M.show()
```

```python
#Initiating the Agglomerative Clustering 
AC = AgglomerativeClustering(n_clusters=4)

# Predict n clusters
yhat_AC = AC.fit_predict(PCA_ds)
PCA_ds["Clusters"] = yhat_AC

#Including Clusters as a feature to the orignal dataframe.
data["Clusters"]= yhat_AC
```

```python
#Plotting the created clusters

fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_ds["Clusters"], marker='o', cmap = cmap )
ax.set_title("Plot Of The Clusters")
plt.show()
```

```python
#Conducting analysis on the clusters 

pal = ["#99DDCC","#FF5DA2", "#9C19E0","#000D6B"]
pl = sns.countplot(x=data["Clusters"], palette= pal)
pl.set_title("Distribution Of The Clusters")
plt.show()
```
